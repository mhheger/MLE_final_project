```python
import pandas as pd
import matplotlib.pyplot as plt
import time

# Function to calculate means of subsets
def means(x, size=100):
    l = len(x)
    x_means = []
    for i in range(l // size):
        subset_mean = x[i*size:(i+1)*size].mean()
        x_means.append(subset_mean)
    return x_means

# Continuously update plots
while True:
    size = 100
    plt.figure(figsize=(15, 8))
    
    # Read the CSV file
    stats = pd.read_csv("/Users/manuelheger/Documents/Organisatorisches Uni/Semester 6/Vorlesungen/Machine Learning Essentials/Abschlussprojekt/bomberman_rl/bomberman_rl/agent_code/renate/stats.csv")
    
    # Calculate means for each metric
    loss = means(stats['mean_loss'], size)
    coins = means(stats['coins_collected'], size)
    crates = means(stats['crates_destroyed'], size)
    opponents = means(stats['opponents_killed'], size)
    steps = means(stats['steps_survived'], size)
    invalid = means(stats['invalid_actions'], size)
    score = means(stats['score'], size)
    reward = means(stats['total_reward'], size)
    
    # Plot all metrics in a 2x4 grid
    plt.subplot(2, 4, 1)
    plt.plot(loss)
    plt.title("Mean Loss")

    plt.subplot(2, 4, 2)
    plt.plot(coins)
    plt.title("Coins Collected")

    plt.subplot(2, 4, 3)
    plt.plot(crates)
    plt.title("Crates Destroyed")

    plt.subplot(2, 4, 4)
    plt.plot(opponents)
    plt.title("Opponents Killed")

    plt.subplot(2, 4, 5)
    plt.plot(steps)
    plt.title("Steps Survived")

    plt.subplot(2, 4, 6)
    plt.plot(invalid)
    plt.title("Invalid Actions")

    plt.subplot(2, 4, 7)
    plt.plot(score)
    plt.title("Score")

    plt.subplot(2, 4, 8)
    plt.plot(reward)
    plt.title("Total Reward")

    # Display the plot
    plt.tight_layout()
    plt.show()
    
    # Pause for 60 seconds
    time.sleep(60)
```
